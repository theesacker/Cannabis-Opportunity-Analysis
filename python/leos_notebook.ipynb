```python
import pandas as pd

county_sales_csv = "resources/Cannabis_Sales_By_County.csv"
census_pop_csv = "resources/Census_pop.csv"
county_per_cap_income_csv = "resources/County Per capita income.csv"
retail_licenses_csv = "resources/retail-licenses.csv"
```

```python
county_sales_df = pd.read_csv(county_sales_csv)
census_pop_df = pd.read_csv(census_pop_csv)
county_per_cap_income_df = pd.read_csv(county_per_cap_income_csv)
retail_licenses_df = pd.read_csv(retail_licenses_csv)


county_sales_df
```

```python
census_pop_df
```

```python
oregon_city_pop = census_pop_df.loc[census_pop_df["Geographic Area"].str.endswith("Oregon"), ["Geographic Area", "Census", "2019"]].reset_index(drop=True)
oregon_city_pop = oregon_city_pop.rename(columns={"Census": "2020"})
oregon_city_pop
```

```python
county_per_cap_income_df
```

```python
for index, row in county_per_cap_income_df.iterrows():
    county_name = row["County"].split(", ")[0]
    
    county_per_cap_income_df.loc[index, "County"] = county_name
    
county_per_cap_income_df
```

```python
retail_licenses_df
```

```python
retail_licenses_df["COUNTY"].unique()
```

```python
dispo_by_county = retail_licenses_df.groupby("COUNTY").agg({"TRADE NAME": "count"}).reset_index(drop=False).rename(columns={"COUNTY": "County", "TRADE NAME": "Dispensary Count"})
dispo_by_county["County"] = dispo_by_county["County"].str.title()
dispo_by_county
```

```python
first_merge = county_per_cap_income_df.merge(dispo_by_county, on="County", how="outer")
```

```python
county_clean_merge = first_merge.dropna().reset_index(drop=True)
county_clean_merge
```

```python
sorted_county = county_clean_merge.sort_values("Personal income (thousands of dollars)", ascending=False)
sorted_county["Population per Dispenary"] = round(sorted_county["Population"] / sorted_county["Dispensary Count"], 0)
sorted_county
```
